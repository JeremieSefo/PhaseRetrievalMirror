## Comparison of discrete and analytic Fourier Transform (Jeremie Sefo)
We compare discrete and analytic Fourier Transforms. We do this for a Gaussian and a square around the origin, and a phantom with those objects replicated along the coordinate axes.

Some notes: The approximation is fairly exact, provided we restrict ourselves to values of the Fourier Transform which are covered by the sampling theorem. Note that there will be windowing effects once we approach the border. Also note that the errors, as can simply be proved, are mostly in the high Fourier coefficients. This implies that for a Gaussian source, there will be almost no errors at all, but for an object with discontinuities, the error will be large. See the squares example!

Note to Jeremy: Be careful when only using Gaussians as your objects. Gaussians and their Fourier Transform decay exponentially towards infinity, however this nice property will not generally be true. Once you have discontinuities like in the square, the Fourier Transform will be much wider. This might help or be a problem. However, both should be tested.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 156  # Grid size
x = np.linspace(-N//2, N//2, N)
y = np.linspace(-N//2, N//2, N)
X, Y = np.meshgrid(x, y)
xoff=0 #4.2
yoff=0 #4.2
# Define the Gaussian function
sigma = 11.1
f = np.exp(-((X-xoff)**2 + (Y-yoff)**2) / (2 * sigma**2))

# Compute the discrete Fourier transform
fhat_discrete = np.fft.fftshift(np.fft.fft2(f))
# fhat_discrete /= np.max(np.abs(fhat_discrete))  # Normalize for comparison
freqs = np.fft.fftfreq(fhat_discrete.shape[0], d = 1)

# Analytical Fourier Transform of the Gaussian
k = np.linspace(freqs.min(), freqs.max(), 1* N) #* (1 / (x[1] - x[0]))  # Frequency domain scaling
KX, KY = np.meshgrid(k, k)
fhat_analytical = np.exp(-0.5 * sigma**2 * (KX**2 + KY**2))



# Plot results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original Gaussian function
im = axes[0].imshow(f, extent=[x.min(), x.max(), y.min(), y.max()], cmap='viridis', origin='lower')
axes[0].set_title("Original Gaussian f(x)")
plt.colorbar(im, ax=axes[0])


# Plot analytical Fourier Transform
im = axes[1].imshow((np.abs(fhat_analytical)), extent=[k.min(), k.max(), k.min(), k.max()], cmap='magma', origin='lower')
axes[1].set_title("Analytical Fourier Transform (|F(k)|)")
plt.colorbar(im, ax=axes[1])

# Plot discrete Fourier Transform
im = axes[2].imshow((np.abs(fhat_discrete)), extent=[k.min(), k.max(), k.min(), k.max()], cmap='magma', origin='lower')
axes[2].set_title("Discrete Fourier Transform (|DFT|)")
plt.colorbar(im, ax=axes[2])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 256  # Grid size
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
X, Y = np.meshgrid(x, y)

# Define the Gaussian function parameters
sigma = 0.1
shift = 3.8  # Distance from the origin for shifted Gaussians

# Define the function f(x) as four Gaussians
f = (np.exp(-((X - shift)**2 + Y**2) / (2 * sigma**2)) +
     np.exp(-((X + shift)**2 + Y**2) / (2 * sigma**2)) +
     np.exp(-(X**2 + (Y - shift)**2) / (2 * sigma**2)) +
     np.exp(-(X**2 + (Y + shift)**2) / (2 * sigma**2)))

# Define frequency domain for Fourier Transform
k = np.linspace(-np.pi, np.pi, N) * (1 / (x[1] - x[0]))  # Frequency scaling
KX, KY = np.meshgrid(k, k)

# Compute the analytical Fourier transform using the shift theorem
fhat_analytical = np.exp(-0.5 * sigma**2 * (KX**2 + KY**2)) * (
    np.exp(-1j * KX * shift) + np.exp(1j * KX * shift) + 
    np.exp(-1j * KY * shift) + np.exp(1j * KY * shift)
)

# Convert to real magnitude (log scale)
fhat_analytical_magnitude = np.log1p(np.abs(fhat_analytical))

# Compute the discrete Fourier transform
fhat_discrete = np.fft.fftshift(np.fft.fft2(f))
fhat_discrete /= np.max(np.abs(fhat_discrete))  # Normalize for comparison

# Plot results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original Gaussian function
axes[0].imshow(f, extent=[-5, 5, -5, 5], cmap='viridis', origin='lower')
axes[0].set_title("Four Gaussians f(x)")

# Plot corrected analytical Fourier Transform
axes[1].imshow(fhat_analytical_magnitude, extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='magma', origin='lower')
axes[1].set_title("Corrected Analytical Fourier Transform log(|F(k)|)")

# Plot discrete Fourier Transform
axes[2].imshow(np.log1p(np.abs(fhat_discrete)), extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='magma', origin='lower')
axes[2].set_title("Discrete Fourier Transform log(|DFT|)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
N = 256  # Grid size
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
X, Y = np.meshgrid(x, y)

# Define the characteristic function of four squares
L = 0.1  # Side length of each square
D = 4.0  # Distance from the origin

f = (np.where((np.abs(X - D) <= L/2) & (np.abs(Y) <= L/2), 1, 0) +
     np.where((np.abs(X + D) <= L/2) & (np.abs(Y) <= L/2), 1, 0) +
     np.where((np.abs(X) <= L/2) & (np.abs(Y - D) <= L/2), 1, 0) +
     np.where((np.abs(X) <= L/2) & (np.abs(Y + D) <= L/2), 1, 0))

# Define frequency domain for Fourier Transform
k = np.linspace(-np.pi, np.pi, N) * (1 / (x[1] - x[0]))  # Frequency scaling
KX, KY = np.meshgrid(k, k)

# Compute the analytical Fourier transform (sinc function with phase shifts)
def sinc(x):
    return np.sinc(x / np.pi)

fhat_analytical = (L**2 * sinc(L * KX / 2) * sinc(L * KY / 2) * 
                   (np.cos(D * KX) + np.cos(D * KY)))

# Convert to real magnitude (log scale)
fhat_analytical_magnitude = np.log1p(np.abs(fhat_analytical))

# Compute the discrete Fourier transform
fhat_discrete = np.fft.fftshift(np.fft.fft2(f))
fhat_discrete /= np.max(np.abs(fhat_discrete))  # Normalize for comparison

# Plot results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot original characteristic function
axes[0].imshow(f, extent=[-5, 5, -5, 5], cmap='viridis', origin='lower')
axes[0].set_title("Characteristic Function of Four Squares f(x)")

# Plot analytical Fourier Transform
axes[1].imshow(fhat_analytical_magnitude, extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='magma', origin='lower')
axes[1].set_title("Analytical Fourier Transform log(|F(k)|)")

# Plot discrete Fourier Transform
axes[2].imshow(np.log1p(np.abs(fhat_discrete)), extent=[-np.pi, np.pi, -np.pi, np.pi], cmap='magma', origin='lower')
axes[2].set_title("Discrete Fourier Transform log(|DFT|)")

plt.tight_layout()
plt.show()
```
