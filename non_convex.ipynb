```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = 5
a = 1
def f(x):
    return 0.25 * ( a**2 * x**2 - y**2)**2
def gradf(x):
    return a**2 * x * ( a**2 * x**2 - y**2)
def hessf(x):
    return a**2 *  ( 3 * a**2 * x**2 - y**2)
```

```python
x = np.linspace(-8, 8)
y1 = f (x)
y2 = gradf(x)
y3 = hessf(x)
'''
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot on the first subplot
ax1.plot(x, y1)
ax1.set_title('f(x) = 0.25 * (y**2 - a**2 * x**2 )**2')

# Plot on the second subplot
ax2.plot(x, 2 * y)
ax2.set_title('Cosine Wave')

# Adjust layout
plt.tight_layout()
plt.show()
'''


plt.plot(x, y1, label = "f(x) = 0.25 * (y**2 - a**2 * x**2 )**2")
plt.plot(x, y2, label = "f'(x) = a**2 * x * ( a**2 * x**2 - y**2)")
plt.plot(x, y3, label = "f''(x) = a**2 *  ( 3 * a**2 * x**2 - y**2)")
#plt.title("f(x) = 0.25 * (y**2 - a**2 * x**2 )**2")
plt.legend(loc = 'upper right')
plt.show()
```
